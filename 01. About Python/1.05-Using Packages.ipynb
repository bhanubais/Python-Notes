# Using Packages

## Installing Packages

- Ensure pip, setuptools, and wheel are up to date
    + `python -m pip install --upgrade pip setuptools wheel`


- Install Package using pip
    + `pip install "SomeProject"`
- Install a Specific Version
    + `pip install "SomeProject==1.4"`
- Install a version that is **compatible** with a certain version
    + `pip install "SomeProject~=1.4.2"`
- Install greater than or equal to one version and less than another:
    + `pip install "SomeProject>=1,<2"`

You can see available Python Packages [Here](https://pypi.org/project/scipy/)

## Numpy Array

```python
from numpy import array

a = array([4, 3, 10, 15])
print(repr(a))

print(repr(a + 2))
print(repr(a + a))
print(repr(a ** 2))
```

## Plotting a Simple Graph

```python
# Plotting Graph inside Current Jupyter Notebook
%matplotlib inline

from matplotlib.pyplot import plot

# Simple exponential Graph f(x) = x**2
plot(a, a**2)
```

```python
# Plotting a dataset
data = [[2,4,6,8,10], [10,15,5,35, 8]]
plot(data[0], data[1])
```

```python
# Plot Numpy Arrays

from numpy import array
a = array([4, 3, 10, 15])
b = array([10, 0.3, 40, 25])

plot(a, b)
```
